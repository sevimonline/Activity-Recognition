# Import Library

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
columnname = pd.read_excel("features.xlsx")
column = columnname["degiskenismi"]
```

```python
x_train = pd.read_excel("final_x_train.xlsx", names = column)
x_test = pd.read_excel("final_x_test.xlsx", names = column)
y_train = pd.read_excel("final_y_train.xlsx", names = ["activity"]) - 1
y_test = pd.read_excel("final_y_test.xlsx", names = ["activity"]) - 1
```

```python
x_train.head()
```

```python
y_train.value_counts()
```

```python
## MOVING ACTIVIES
y_train = y_train.replace(0,1)
y_train = y_train.replace(1,1)
y_train = y_train.replace(2,1)

y_test = y_test.replace(0,1)
y_test = y_test.replace(1,1)
y_test = y_test.replace(2,1)


## STATIONARY ACTIVITIES
y_train = y_train.replace(3,0)
y_train = y_train.replace(4,0)
y_train = y_train.replace(5,0)

y_test = y_test.replace(3,0)
y_test = y_test.replace(4,0)
y_test = y_test.replace(5,0)
```

```python
y_train.value_counts()
```

```python
from lazypredict.Supervised import LazyClassifier
```

# Model with 'accelerometer' and 'gyroscope' sensors.

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(x_train, x_test, y_train, y_test)
models
```
